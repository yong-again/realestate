```python
def rename_col(df1, df2, df3, df4, df5, df6, df7, df8, df9 ,col1, col2):
    df1.rename(columns={col1:col2}, inplace=True)
    df2.rename(columns={col1:col2}, inplace=True)
    df3.rename(columns={col1:col2}, inplace=True)
    df4.rename(columns={col1:col2}, inplace=True)
    df5.rename(columns={col1:col2}, inplace=True)
    df6.rename(columns={col1:col2}, inplace=True)
    df7.rename(columns={col1:col2}, inplace=True)
    df8.rename(columns={col1:col2}, inplace=True)
    df9.rename(columns={col1:col2}, inplace=True)
```
